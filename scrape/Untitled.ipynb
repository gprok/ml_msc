```python
from textblob import TextBlob
```

```python
import nltk
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
text = "The new product is terrible. I hated every feature of it!"
```

```python
scores = sia.polarity_scores(text)
```

```python
scores
```

```python
score = scores['compound']
if score >= 0.05:
    print("Positive")
elif score <= -0.05:
    print("Negative")
else:
    print("Neutral")
```

```python
blob = TextBlob(text)
```

```python
polarity = blob.sentiment.polarity
```

```python
polarity
```

```python
if polarity >= 0.05:
    print("Positive")
elif polarity <= -0.05:
    print("Negative")
else:
    print("Neutral")
```
