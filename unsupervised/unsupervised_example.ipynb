```python
from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [amount spent, number of visits]
customers = np.array([
    [20, 2], [47, 3], [400, 15], [187, 12], [32, 2], [240, 17], [100, 5], [96, 7], [1500, 1], [400, 22]
])
```

```python
km = KMeans(n_clusters=5, random_state=42)
```

```python
groups = km.fit_predict(customers)
```

```python
groups
```

```python
plt.scatter(customers[:, 0], customers[:, 1], c=groups)
plt.xlabel('Amount Spent')
plt.ylabel('Number of visits')
plt.title('Customer Groups Identified')
```
